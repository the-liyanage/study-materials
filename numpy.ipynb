```python
#Numerical python
#ndarrays -----> n- dimentional arrays

import numpy as np
```

```python
from time import process_time
```

```python
#Time taken by a python list

my_list = [i for i in range(10000)]
start_time = process_time()
my_list = [i+5 for i in my_list]
end_time = process_time()

#Overall time taken for the process

overall_time_list = end_time - start_time
print(overall_time_list)
```

```python
#Time taken by an ndarray

my_array = np.array([i for i in range(10000)])
start_time = process_time()
my_array += 5
end_time = process_time()

#Overall time taken for the process

overall_time_array = end_time - start_time
print(overall_time_array)
```

```python
#Comparing the time components of each other

if overall_time_array > overall_time_list:
    print("Arrays take a greater time to process than the lists")

else:
    print("Arrays take a lesser time to process than the lists")
```

```python
#Numerical python
#ndarrays -----> n- dimentional arrays

import numpy as np
```

```python
from time import process_time
```

```python
#Time taken by a python list

my_list = [i for i in range(10000)]
start_time = process_time()
my_list = [i+5 for i in my_list]
end_time = process_time()

#Overall time taken for the process

overall_time_list = end_time - start_time
print(overall_time_list)
```

```python
#Time taken by an ndarray

my_array = np.array([i for i in range(10000)])
start_time = process_time()
my_array += 5
end_time = process_time()

#Overall time taken for the process

overall_time_array = end_time - start_time
print(overall_time_array)
```

```python
#Comparing the time components of each other

if overall_time_array > overall_time_list:
    print("Arrays take a greater time to process than the lists")

else:
    print("Arrays take a lesser time to process than the lists")
```

```python
np1 = np.array([1,2,3,4,5])
print(np1)
print()
#length of the array 
print(f"The length of the array is\n{np1.shape}\n")


#Range of numbers
np2 = np.arange(10)
print(f"The number of elements of an array given by range(10)\n{np2}\n")


#Range of numbers with a step-size
np3 = np.arange(1,12,2)
print(f"The number of elements of an array with a step-size of 2\n{np3}\n")


#Arrays with zeros as elements
np4 = np.zeros(10)
print(np4)
print()

#Creating two-dimensional arrays
np_two = np.array([(1,2,3), (4,4,6)])
print(f"This is a two-dimensional array\n {np_two}")
print(f"This is the shape of the 2-dimensional array {np_two.shape}\n")


#Creating float-arrays
np_flt = np.array([(2,3,4), (5,6,7)], dtype='float')
print(f"This is an array with float elements\n{np_flt}\n")


#Mutidimensional zeros array
#2 dimensional array with 7 elements
np5 = np.zeros((2,7))
print(np5)
print()

#full arrays
# five elements with the same number(4)
np6 = np.full((5),4)
print(np6)
print()

#Mutidimensional full arrays
np7 = np.full((2,5),3)
print(np7)
print()

#Converting a list to an array
my_list = [1,2,4,5,6]
my_array = np.array(my_list)
print(my_array)
```

```python
#Slicing Numpy arrays 
np1 = np.array([1,2,3,4,5,6,7,8])

#Return 2,3,4,5
print(np1[1:5])
print()

#Return from something till the end
print(np1[3:])
print()

#Return negative slices
print(np1[-3:-1])
print()

```
